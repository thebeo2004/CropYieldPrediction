```python
import pandas as pd 
import numpy as np
from osgeo import gdal
```

Define a dataframe with numerous attributes.

```python
months = ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X', 'XI', 'XII']
collumns = ['Năm', 'TT']

for month in months:
    collumns.append('NDVI_' + month)
```

```python
table = pd.DataFrame()

for i in range(len(collumns)):
    table.insert(i, collumns[i], None)
```

```python
table
```

Extract NDVI from dataset and write them into constructed dataframe 

```python
provinces = ['Hai_Phong', 'Hung_Yen', 'Nam_Dinh', 'Ninh_Binh', 'Quang_Ninh', 'Thai_Binh', 'Vinh_Phuc', 'Bac_Ninh', 'Ha_Nam', 'Ha_Noi', 'Hai_Duong']
folder_in = "Data\\NDVI of Rice Paddies\\"
```

```python
def extract_NDVI(file_in):
    dataset = gdal.Open(file_in, gdal.GA_ReadOnly)
    NDVI_band = dataset.GetRasterBand(1)
    band_min, band_max, band_mean, band_stddev = NDVI_band.GetStatistics(0, 1)

    return band_mean
```

```python
for province in provinces:
    for year in range(2000, 2021):

        row = [year, province]

        for month in range(1, 13):

            if (year == 2000 and month == 1):
                row.append(None)
                continue

            str_year = str(year)
            str_month = str(month)

            if (month < 10):
                str_month = '0' + str_month

            file_in = folder_in + "\\" + province + "\\" + str_year + "\\" + str_month + ".tif"

            row.append(extract_NDVI(file_in=file_in))
            
        # print(row)
        table.loc[len(table)] = row
```
