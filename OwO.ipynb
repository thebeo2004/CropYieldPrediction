```python
import numpy as np
import pandas as pd
```

```python
ndvi_df = pd.read_csv(r'Data\NDVI.csv')
ndvi_df.head()
```

```python
ndvi_df.columns
```

```python
provinces = ['Ha_Noi', 'Hai_Duong', 'Hung_Yen', 'Nam_Dinh', 'Ninh_Binh', 'Thai_Binh', 'Ha_Nam']
```

```python
ndvi_df = ndvi_df.loc[ndvi_df['TT'].isin(provinces)]
ndvi_df
```

```python
previous_fields = ['NDVI_X', 'NDVI_XI', 'NDVI_XII']
current_fields = ['NDVI_I', 'NDVI_II', 'NDVI_III', 'NDVI_IV', 'NDVI_V']
```

```python
columns = ['Năm', 'TT']

for field in previous_fields:
    columns.append(field)

for field in current_fields:
    columns.append(field)
```

```python
table = pd.DataFrame()

for i in range(len(columns)):
    table.insert(i, columns[i], None)
```

```python
table
```

```python
for province in provinces:
    for year in range(2001, 2021):
        row = [year, province]

        for field in previous_fields:
            tmp = ndvi_df.loc[(ndvi_df['TT'] == province) & (ndvi_df['Năm'] == (year - 1))]
            row.append(list(tmp[field])[0])
        
        for field in current_fields:
            tmp = ndvi_df.loc[(ndvi_df['TT'] == province) & (ndvi_df['Năm'] == (year))]
            row.append(list(tmp[field])[0])
        
        table.loc[len(table)] = row
```

```python
table.to_csv('NDVI.csv', index=False)
```
