```python
from sklearn.linear_model import RidgeCV
from SplittingData import splitting_data
from Metrics import metrics_table, metrics_calculation
```

```python
X_train, y_train, test_set = splitting_data()
```

```python
ridge_cv = RidgeCV(alphas=[0.1, 1, 2, 5, 9, 10, 20, 50, 100])
ridge_cv.fit(X_train, y_train)
```

```python
print("Ridge alpha:", ridge_cv.alpha_)
print("Ridge coefficients:", ridge_cv.coef_)
```

```python
avg_y_train=y_train['RandomCropYield'].mean()
```

```python
print("Metrics calculation on Training Set:\n", metrics_calculation(y_actual=y_train, y_pred=ridge_cv.predict(X_train), avg_y_train=avg_y_train))
print("Metrics calculation on Test Set (From 2017 - 2020):")

metrics_table(starting_year=2017, test_set=test_set, avg_y_train=avg_y_train, estimator=ridge_cv)
```
