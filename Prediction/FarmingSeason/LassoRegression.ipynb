```python
from sklearn.linear_model import LassoCV
from SplittingData import splitting_data
from Metrics import metrics_calculation, metrics_table
```

```python
X_train, y_train, test_set = splitting_data()
```

```python
lasso_cv = LassoCV(max_iter=5000, cv=7, n_jobs=-1)
lasso_cv.fit(X_train, y_train)
```

```python
print("Lasso alpha:", lasso_cv.alpha_)
print("Lasso coefficients:", lasso_cv.coef_)
```

```python
avg_y_train=y_train['RandomCropYield'].mean()
```

```python
print("Metrics calculation on Training Set:\n", metrics_calculation(y_actual=y_train, y_pred=lasso_cv.predict(X_train), avg_y_train=avg_y_train))
print("Metrics calculation on Test Set (From 2017 - 2020):")

metrics_table(starting_year=2017, test_set=test_set, avg_y_train=avg_y_train, estimator=lasso_cv)
```
